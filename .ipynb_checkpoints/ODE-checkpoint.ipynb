```python
###sympy for symbolic comp
from sympy import *
```

```python
##not as in matlab
##dsolve('Dy-y=exp(-t)*cos(3*t)','y(0)=1')
## rewritten as Dy-y-exp(-t)*cos(3*t)
init_printing(use_latex=True) 
var('t C1')
y=Function('y')(t)
de=Eq(y.diff(t)-y-exp(-t)*cos(3*t))
display(de)
```

```python
des=dsolve(de,y)
display(des)
```

```python
des.subs(C1,0)
```
